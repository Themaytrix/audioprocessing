```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading music files
debussy_file = "audio/debussy.wav"
debusy,sr = librosa.load(debussy_file)
```

### Mel Filter Banks

Mel filter banks help get blocks of mel bands of frequencies that are perceptually relevant 


```python
filter_banks = librosa.filters.mel(n_fft=2048,sr=sr,n_mels=10)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(filter_banks,
                        sr=sr,
                        x_axis="linear")
plt.colorbar(format="%+2.f")
```

### Extracting Mel Spectogram

```python
# returns linear mel_spec
mel_spectrogram = librosa.feature.melspectrogram(y=debusy,sr=sr,n_mels=90)
```

```python
# convert to logarithmic val
log_mel_spectogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectogram,
                         x_axis="time",
                         y_axis='log')
plt.colorbar(format="%+2.f")
```
